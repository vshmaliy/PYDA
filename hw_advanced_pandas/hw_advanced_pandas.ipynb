### Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

- если источник traffic_source равен yandex или google, то в source_type ставится organic
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep=';')
log.head(20)
```

```python
providers = ['yandex','google']
other_sources = ['paid','email']
```

```python
log['source_type'] = log['traffic_source']
for provider in providers:
    log.loc[log.traffic_source == provider, 'source_type'] = 'organic'
for source in other_sources:
    log.loc[((log.traffic_source == source)) & (log.region=='Russia'), 'source_type'] = 'ad'
    log.loc[((log.traffic_source == source)) & (log.region!='Russia'), 'source_type'] = 'other'  
log.head(20)
```

### Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

#### Примечание:
Я интерпретировал задание так, что среднее время жизни пользователей - это среднее время жизни по множеству пользователей, которые выставили более 100 оценок. Соответственно, я сначала получаю таблицу пользователей которые выставили более 100 оценок и создаю в ней колонку с временем жизни для каждого такого userId, и после этого считаю среднее время жизни на основании этой таблицы.

```python
import pandas as pd
```

```python
ratings=pd.read_csv('ml-latest-small/ratings.csv', sep=',')
ratings.head()
```

```python
ratings=ratings[['userId','timestamp']]
ratings.head()
```

```python
users=ratings.groupby('userId').count()
users_100_plus = users[users['timestamp']>100].reset_index()
users_100_plus = users_100_plus[['userId']]
users_100_plus.head()
```

```python
for i in users_100_plus.userId:
    ratings.loc[ratings.userId == i, 'lifetime'] = (ratings[ratings.userId ==i].timestamp.max()-ratings[ratings.userId ==i].timestamp.min())
```

```python
joined = users_100_plus.merge(ratings, on='userId', how='left')
joined=joined[['userId','lifetime']].drop_duplicates().reset_index()
joined=joined[['userId','lifetime']]
joined.head()
```

```python
average_lifetime=round(joined.lifetime.sum()/(joined.index.max()+1))
```

```python
from datetime import timedelta
sec = average_lifetime
td = timedelta(seconds=sec)
print('Average lifetime of users with more than 100 reviews is',average_lifetime, 'sec, or', td)
```

### Задание 4

Дана статистика услуг перевозок клиентов компании по типам:
- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

Необходимо сформировать две таблицы:
- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
- аналогичную таблицу по типам выручки с указанием адреса клиента

*Обратите внимание, что в процессе объединения таблиц данные не должны теряться.*

```python
import pandas as pd
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

### Вариант 1
Ключевая база - клиентская, т.к. она содержит все client_id
Поэтому можно начать с построения полной базы, и затем удалить адреса

```python
table=client_base.merge(rzd, how='left', on='client_id')
table=table.merge(auto, how='left', on='client_id')
table=table.merge(air, how='left', on='client_id')
table
```

```python
#Для удобочитаемости можно заменить NaN нулями и заменить индекс на client_id

table.loc[table.rzd_revenue.isnull(), 'rzd_revenue',]=0
table.loc[table.auto_revenue.isnull(), 'auto_revenue',]=0
table.loc[table.air_revenue.isnull(), 'air_revenue',]=0
table_addr=table.set_index('client_id')
table_addr
```

```python
# И уже из этой таблицы создаем таблицу с тремя типами выручки для каждого client_id без указания адреса клиента

table_no_addr=table_addr.drop(['address'], axis=1)
table_no_addr
```

### Вариант 2
Можно объединить таблицы rzd, auto, air через outer join и затем полученную так таблицу через left join присоединить к client_base

```python
table=rzd.merge(auto, how='outer', on='client_id')
table=table.merge(air, how='outer', on='client_id')
table.loc[table.rzd_revenue.isnull(), 'rzd_revenue',]=0
table.loc[table.auto_revenue.isnull(), 'auto_revenue',]=0
table.loc[table.air_revenue.isnull(), 'air_revenue',]=0
table=table.set_index('client_id')
table
```

```python
table_addr=client_base.merge(table, how='left', on='client_id')
table_addr=table_addr.set_index('client_id')
table_addr
```
