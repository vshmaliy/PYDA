### Задание 1

Скачайте с сайта grouplens.org...movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies = pd.read_csv('ml-latest-small/movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
rated5 = joined[ joined.rating == 5.0 ]
rated5.head()
```

```python
rated5.title.value_counts().head(10)
```

Итак, в датасете ml-latest-small больше всего оценок 5.0 было выставлено фильму The Shawshank Redemption (1994).

Впрочем, весь top-10 впечатляет - интересно, что при использовании датасета ml-25m первая десятка остается той же - вариации начинаются где-то в третье десятке. Я в конечном итоге использовал меньший датасет из экологичесаких соображений)

### Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
baltics = data[ 
               ((data['country']=='Lithuania') | (data['country']=='Latvia')| (data['country']=='Estonia'))&
               ((data['category']==4) | (data['category']==12) | (data['category']==21))&
               (data['year']>2005)&(data['year']<2010)&
               (data['quantity']>0)]
baltics.head()
```

```python
print("Суммарное потребление стран Прибалтики категорий 4, 12, 21 с 2005 по 2010 год:", baltics['quantity'].sum())
```

```python
#на всякий случай решил перепроверить в экселе - все сошлось)
baltics.to_excel("baltics.xlsx")
```
