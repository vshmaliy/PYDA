````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
````

````python
def func_p():
    """
    Эта функция позволяет узнать владельца документа по его номеру
    """
    doc_num = (input('Введите номер документа'))
    name_num_dict={}
    for doc in documents:
        name_num_dict[doc['number']] = doc['name']
    if doc_num in list(name_num_dict.keys()):
        print ('Владелец документа:',name_num_dict[doc_num])
    else:
        print ('Номер документа отсутствует или введен некорректно')

def func_s():
    """
    Эта функция позволяет по номеру документа узнать на какой полке он хранится
    """
    doc_num = (input('Введите номер документа'))
    shelf_num_dict={}
    for directory in directories:
        for docs in list(directories[directory]):
            shelf_num_dict[docs] = directory
    if doc_num in list(shelf_num_dict.keys()):
        print ('Документ хранится на полке:',shelf_num_dict[doc_num])
    else:
        print ('Номер документа отсутствует или введен некорректно')

def func_l():
    """
    Эта функция позволяет получить полную информацию по всем документам
    """
    for directory, docs in directories.items():
        for doc in documents:
            if doc['number'] in docs:
                print(f"№{doc['number']}, тип: {doc['type']}, враделец: {doc['name']}, полка хранения: {directory}")
def func_ads():
    """
    Эта функция позволяет добавить папку
    """
    directory = (input('Введите номер полки'))
    if directory in directories:
        print('Такая полка уже существует. Текущий перечень полок:', ', '.join(list(directories.keys())))
    else:
        directories[directory] = []
        print('Полка добавлена. Текущий перечень полок:', ', '.join(list(directories.keys())))

def func_ds():
    """
    Эта функция позволяет удалить папку
    """
    directory = (input('Введите номер полки'))
    if directory in directories:
        if len(directories[directory]) == 0:
            del(directories[directory])
            print('Полка удалена. Текущий перечень полок:', ', '.join(list(directories.keys())))
        else:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:', ', '.join(list(directories.keys())))
    else:
        print('Такой полки не существует. Текущий перечень полок:', ', '.join(list(directories.keys())))

def func():
    """
    Эта функция включает в себя совокупность функций, позволяющих выбрать следующие варианты команд:
    
    - узнать владельца документа по его номеру,
    - узнать по номеру документа, на какой полке он хранится,
    - получить полную информацию по всем документам,
    - добавить папку,
    - удалить папку.
    """
    choice = (input('Введите команду:'))
    if choice == 'p':
        func_p()
    elif choice == 's':
        func_s()
    elif choice == 'l':
        func_l()
    elif choice == 'ads':
        func_ads()
    elif choice == 'ds':
        func_ds()
    else:
        print('Такой команды нет.')
        print('Вы можете использовать следующие команды:')
        print('p - узнать владельца документа по его номеру')
        print('s - по номеру документа узнать на какой полке он хранится')
        print('l - получить полную информацию по всем документам')
        print('ads - добавить папку')
        print('ds - удалить папку')
````
